```python
df['label'] = df['sentiment'].map({'positive': 1, 'negative': 0})
```

```python
df = df[['review', 'label']]
df.head()
```

```python
def remove_html(text):
    return BeautifulSoup(text, "html.parser").get_text()
```

```python
def remove_special_chars(text):
    return re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
def to_lower(text):
    return text.lower()
```

```python
def remove_html(text):
    return BeautifulSoup(text, "html.parser").get_text()
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    tokens = text.split()
    filtered = [word for word in tokens if word not in stop_words]
    return " ".join(filtered)
```

```python
stemmer = PorterStemmer()

def apply_stemming(text):
    tokens = text.split()
    stemmed = [stemmer.stem(word) for word in tokens]
    return " ".join(stemmed)
```

```python
df['clean_review'] = df['review'].apply(remove_html)
df['clean_review'] = df['clean_review'].apply(remove_special_chars)
df['clean_review'] = df['clean_review'].apply(to_lower)
df['clean_review'] = df['clean_review'].apply(remove_stopwords)
df['clean_review'] = df['clean_review'].apply(apply_stemming)
```
